```python
#Supprimer la colonne "Valeur"
df.drop(['Valeur'],inplace=True,axis=1)
```

```python
#Afficher les statistiques descriptives pour les variables quantitatives de l'objet df
df.select_dtypes(include = ['int','float']).describe()
```

```python
#Afficher les statistiques descriptives pour les variables qualitatives de l'objet df
df.select_dtypes(include='object').describe()
```

```python
#On veut afficher les produits dont la catégorie = boissons
df[df['Categorie']=='Boissons']
```

```python
#On veut afficher les produits dont la Catégorie = boissons et le prix >100
df[(df['Categorie']=='Boissons') & (df['Prix']>100)]
```

```python
#On veut afficher les produits dont la Catégorie = boissons et l'origine=CEE et prix > 100
df[(df['Categorie']=='Boissons')& (df['Origine']=='CEE')& (df['Prix']>100)]
```

```python
#On veut afficher les produits dont la (Catégorie = boissons et origine = CEE) OU (catégorie = condiment)
df[(df['Categorie']=='Boissons')&(df['Origine']=='CEE') |(df['Categorie']=='Condiments')]
```

```python
#On veut afficher les produits dont (Catégorie = viande ET origine = CEE) OU (catégorie = condiment ET origine = extérieur)
df[(df['Categorie']=='Viandes')&(df['Origine']=='CEE') |(df['Categorie']=='Condiments')&(df['Origine']=='Exterieur')]
```

```python
#On veut afficher les produits dont le prix > 70 et prix <=100
df[(df['Prix']>70)&(df['Prix']<100)]
```

```python
#Lister les aliments dont le prix est compris entre 100 et 200, et qui sont des « viandes »
df[(df['Prix']>100)&(df['Prix']<200)&(df['Categorie']=='Viandes')]
```

```python
#Lister les 15 produits les moins chers
df.groupby(['Prix']).min().head(15)
```

```python
#Quel sont les 5 produit moin chers vendus a Lyon?
df1=df[['Categorie','Ville','Prix']].groupby(['Prix']).min()
df1[(df1['Ville']=='Lyon')].head()
```

```python
#Calculer la moyenne de prix des boissons distribuées à Lyon (pivot_table)
df.pivot_table('Prix','Categorie','Ville').mean()
```

```python
#En utilisant une boucle afficher la representation graphique de toutes les variables quantitative
list_quant=df.select_dtypes(include=['int','float']).columns
plt.figure(figsize=(20,10))
x=1
for valeur in list_quant:
    plt.subplot(1,2,x)
    plt.hist(df[valeur].value_counts())
    x+=1
    
```

```python
#En utilisant une boucle afficher la representation graphique de toutes les variables qualiative
list_qual=df.select_dtypes(include='object').columns
plt.figure(figsize=(20,10))
x=1
for valeur in list_qual:
    plt.subplot(1,4,x)
    df[valeur].value_counts().plot(kind='bar')

    x+=1  
    
```
